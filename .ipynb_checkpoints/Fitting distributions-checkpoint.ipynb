```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.stats
```

```python
df = pd.read_csv('./income_final.csv', encoding='utf8')
```

```python
df
```

```python
df = df[['Mean']]
```

```python
df
```

```python
size = len(df.index)
```

```python
y = df['Mean'].tolist()
```

```python
x = scipy.arange(size)
```

```python
x
```

```python
n, b, p = plt.hist(y, bins=30, color='y')
sum_y = np.sum(n)*(b[1]-b[0])
new_n = n / sum_y
```

```python
dist_names = ['norm', 'beta']

for dist_name in dist_names:
    dist = getattr(scipy.stats, dist_name)
    param = dist.fit(y)
    pdf_fitted = dist.pdf(x, *param[:-2], loc=param[-2], scale=param[-1])
    
    plt.figure()
    plt.plot(pdf_fitted, label=dist_name)
    plt.hist(b[1:], weights=new_n, bins=30, color='y')
    plt.legend(loc='upper right')
    plt.show()
```

```python
log_y = np.log(y)
```

```python
plt.hist(log_y, bins=30)
```

```python
mean = np.mean(y)
sigma = np.std(y)
```

```python
mean
```

```python
sigma
```

```python
lognormal= scipy.stats.lognorm([sigma**2],loc=mean)
```

```python
plt.plot(x, lognormal.pdf(x))
```
